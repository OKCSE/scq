```python
#  Single Channel Queuing System (Assignment 6)
#  Simulation & Modeling Sessional
#  Group 3 
#       1. Joseph Diaz (CSE-01406362) | https://github.com/Joseph440G0
#       2. Md Woise Uddin Akbar (CSE-01105875) | https://github.com/OKCSE/
#       3. Raghib Shahriar (CSE - 01206008) | https://github.com/Raghib6
#       4. Hasib Abdullah (CSE-01406316)
#       5. Mohammad Azad  (CSE - 01406365) | https://github.com/jrazad10
#
 
# Importing Packages
 
from scipy.stats import *
 
# Declaring List
 
int_arr_time = []
arr_time = []
service_time = []
tsb = []
tse = []
sys_time = []
idle_time = []
waiting_time = []
 
 
mu = 5.6
for i in range (19):
  n = poisson.rvs(mu)   #Random Number Generation using Poission Distribution 
  int_arr_time.append(n)    #Assigning the random to the int_arr_time list
 
#Default Value Assignment for First Customer
 
int_arr_time.insert(0,0)
arr_time.insert(0,0)
tsb.insert(0,0)
waiting_time.insert(0,0)
idle_time.insert(0,0)
 
#Arrival Time Generation
for i in range (1,20):
    arr_time.append(int_arr_time[i] + arr_time[i-1])  #Formula
 
#Service Time Generation
for i in range (20):
  n = expon.rvs(scale=0.5, loc=5)   #Random Number Generation using Exponential Distribution
  service_time.append(round(n,2))    #Assigning the random number to service_time list 
 
 
 
#TSB & TSE Generation
 
tse.insert(0,(service_time[0] + tsb[0]))  #Calculate first customer TSE first
 
for i in range (1,20):
  if (arr_time[i] > tse[i-1]):
    tsb.insert(i,round(arr_time[i],2))
    tse.insert(i,round((tsb[i] + service_time[i]),2))   #Formula
  else:
    tsb.append(tse[i-1])
    tse.insert(i,round((tsb[i] + service_time[i]),2))     #Formula
 
#Waiting Time Generation
 
for i in range (1,20):
  if (tsb[i] > arr_time[i]):
    waiting_time.insert(i, round((tsb[i] - arr_time[i]),2))     #Formula
  else:
    waiting_time.insert(i,0)
 
#Total Time in System Generation
 
for i in range(20):
  sys_time.insert(i, round((service_time[i] + waiting_time[i]),2))    #Formula
 
#Server Idle Time Generation
 
for i in range (1,20):
  if (tse[i-1] < arr_time[i]):
    idle_time.insert(i,round((arr_time[i] - tse[i-1]),2))     #Formula
  else:
    idle_time.insert(i,0)
 
 # -----------------------------------

#Data Visualization using simple print statements
 
hr_border="------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------"
 
print(hr_border)
print("|  Customer No  |  Interarrival Time  |  Arrival Time   |  Service Time      |  Service Time Begin  |  Service Time End  |  Waiting Time  |  Time Spend in System   |  Server Idle Time  |")
print(hr_border)
 
for i in range (1,21):  #Printing the generated values as a table
  print("|\t",i ,"\t    |\t\t", int_arr_time[i-1], "\t    |\t", arr_time[i-1], "\t\t|\t", service_time[i-1], "\t\t|\t  %1.2f"% (tsb[i-1]), "\t   |\t\t%.2f\t" % (tse[i-1]), "   |\t  %.2f"% (waiting_time[i-1]), "\t|\t\t%.2f"% (sys_time[i-1]), "  \t|\t%.2f"% (idle_time[i-1]), "\t\t|")
 
print(hr_border)
 
#Calculating Sum
int_arr_time_sum = sum(int_arr_time)
service_time_sum = round(sum(service_time),2)
waiting_time_sum = round(sum(waiting_time),2)
sys_time_sum = round(sum(sys_time),2)
idle_time_sum = round(sum(idle_time),2)
 
print("|\tSum: \t","   |\t\t=",int_arr_time_sum,"\t    |\t\t\t|\t=",service_time_sum,"\t|\t\t\t   |\t\t\t    |\t=",waiting_time_sum,"\t|\t\t=",sys_time_sum,"\t|\t=",idle_time_sum,"  \t|")
 
print(hr_border)
customer =20
avg_waiting_time = round(waiting_time_sum/customer,2)
print("Average Waiting Time: ",avg_waiting_time)
avg_sys_time = round(sys_time_sum/customer,2)
print("Average Time Spend in System: ",avg_sys_time)
avg_int_arr_time = round(int_arr_time_sum/(customer-1),2)
print("Average Time Between Arrival: ",avg_int_arr_time)
avg_service_time = round(service_time_sum/customer,2)
print("Average Service Time: ",avg_service_time)
prob_idle_time = round(idle_time_sum/tse[19],2)
print("The probability of the server being idle: ",prob_idle_time)
```
